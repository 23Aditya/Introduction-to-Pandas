# Suppose, I want to compute the following probability:

\begin{equation}
P(class = 'p')
\end{equation}

# Well, I have a sample of 8124 Mushrooms, but I don't have the population. Therefore, I can calculate the estimate of this probability using the sample. And the estimate is given by the relative frequency. The relative frequency of Poissonous mushrooms is given by: 

\begin{equation}
\hat{P}(class = poissonous) = \frac{Frequency(Mushroom = Poissonous)}{Total Number of Mushrooms (Sample Size)}
\end{equation}

\begin{equation}
\hat{P}(class = poissonous) = \frac{Number of rows corresponsing to poissonous mushrooms in the Pandas DataFrame}{Sample Size (Total Number of rows in Pandas DataFrame)}
\end{equation}

# So, how we are going to compute the Frequency of poissonous mushrooms in our data. For that, we have to seperate the data of all the poissonous mushrooms from our Pandas DataFrame. This seperated data will be in the form of another smaller DataFrame. Then we will be computing the total number of rows in the smaller DataFrame. 

# Now, how we are going to do that ? 

# In order to understand this, we first have to understand the concept of Boolean masks. 

# So, what are Booleans Masks ? 

# Boolean Masks are actually the DataFrame of Boolean values (True and False) where True corresponds to some condition being True and False corresponds to the same condition being False. 

# Now, the next concept after slicing of DataFrame from iloc as well as loc, is going to be of Boolean Masks. 

```python
df_columns[0]
```

```python
df['class']
```

```python
boolean_mask = df['class'] == 'p'
```

# Now, as you can that the above DataFrame is a DataFrame of Boolean Values and hence it is called boolean mask. 

# Now, if we pass this boolean mask inside the square brackets of our pandas dataframe:

# df[boolean_mask]

# then, let's see what happens:

```python
poissonous_mushrooms = df[boolean_mask]
```

# we have to find now out of this data that, how many mushrooms are there which have poissonous category so for that, we have a function associated with the DataFrame and that is called shape. 

```python
poissonous_mushrooms.shape[0]
```

# So, finally we were able to find the Frequency(Mushroom = Poissonous) = 3916

# So, we have to compute now the total number of mushrooms (sample size) in our data

```python
sample_size = df.shape[0]
```

```python
sample_size
```

```python
p_class_equals_poissonous = poissonous_mushrooms.shape[0]/df.shape[0]
```

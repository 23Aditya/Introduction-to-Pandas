# Suppose, I want to compute the following probability:

\begin{equation}
P(class = 'p')
\end{equation}

# Well, I have a sample of 8124 Mushrooms, but I don't have the population. Therefore, I can calculate the estimate of this probability using the sample. And the estimate is given by the relative frequency. The relative frequency of Poissonous mushrooms is given by: 

\begin{equation}
\hat{P}(class = poissonous) = \frac{Frequency(Mushroom = Poissonous)}{Total Number of Mushrooms (Sample Size)}
\end{equation}

\begin{equation}
\hat{P}(class = poissonous) = \frac{Number of rows corresponsing to poissonous mushrooms in the Pandas DataFrame}{Sample Size (Total Number of rows in Pandas DataFrame)}
\end{equation}

# So, how we are going to compute the Frequency of poissonous mushrooms in our data. For that, we have to seperate the data of all the poissonous mushrooms from our Pandas DataFrame. This seperated data will be in the form of another smaller DataFrame. Then we will be computing the total number of rows in the smaller DataFrame. 

# Now, how we are going to do that ? 

# In order to understand this, we first have to understand the concept of Boolean masks. 

# So, what are Booleans Masks ? 

# Boolean Masks are actually the DataFrame of Boolean values (True and False) where True corresponds to some condition being True and False corresponds to the same condition being False. 

# Now, the next concept after slicing of DataFrame from iloc as well as loc, is going to be of Boolean Masks.

```python
df['class']
```

```python
boolean_mask = df['class'] == 'p'
```

# Now, as you can that the above DataFrame is a DataFrame of Boolean Values and hence it is called boolean mask. 

# Now, if we pass this boolean mask inside the square brackets of our pandas dataframe:

# df[boolean_mask]

# then, let's see what happens:

```python
poissonous_mushrooms = df[boolean_mask]
```

# we have to find now out of this data that, how many mushrooms are there which have poissonous category so for that, we have a function associated with the DataFrame and that is called shape. 

```python
poissonous_mushrooms.shape[0]
```

# Now, let's try to compute another set of Probabilities. Let's say that we want to compute the following probabilities:

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous)
\end{equation}

# The above Probability is actually the Probability of picking up a mushroom whose cap-shape = x as well as cap-surface = y as well as cap-color = w on condition that the mushroom belongs to poissonous category. 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n)
\end{equation}

# The above probability is actually probability of picking up a mushroom whose gill-spacing = c as well as stalk-surface-below-ring = s on a condition that the mushroom belongs to poissonous category as well as it's spore-print-color = n

# So, lets focus on the first type of Probability: 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous)
\end{equation}

# Let's say that we have an event                                                                              A = (cap-shape=x and cap-surface=y and cap-color=w) and similarly, you have another event B = (class = poissonous)

# it means that your event A is actually the event of picking up of random mushroom whose cap-shape=x and cap-surface=y and cap-color=w furthermore, your event B is actually the event of picking up of random mushroom whose category is poissonous. 

# It means that, we can write: 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous) = P(A|B)
\end{equation}

# we know that, according to the formula for conditional Probability: 

\begin{equation}
P(A|B) = \frac{P(A \cap B)}{P(B)}
\end{equation}

# Therefore, 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous) = \frac{P(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap class = poissonous)}{P(class = poissonous)}
\end{equation}

# So, now we have to find two probabilities:

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap class = poissonous)
\end{equation}

# and 

\begin{equation}
P(class = poissonous)
\end{equation}

# We have aleady found out the probability in the denominator, but we have to find the probability in the numerator: 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap class = poissonous)
\end{equation}

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap class = poissonous) = \frac{Frequency(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap clas=poissonous)}{Sample Size (8124)}
\end{equation}

# Which means that, we need to find out:

# Frequency (cap-shape=x and cap-surface=y and cap-color=w and class=poissonous) = Number of rows in pandas dataframe satisfying all the four conditions simultaniously at the same time

# In order to compute this frequency, we need to create the boolean mask of the four conditions simultaniously. 

```python
boolean_mask = (df['cap-shape'] == 'x') & (df['cap-surface'] == 'y') & (df['cap-color'] == 'w') & (df['class'] == 'p')
```

```python
frequency = df[boolean_mask].shape[0]
```

```python
frequency
```

# Therefore, 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w \cap class = poissonous) = \frac{32}{Sample Size (8124)}
\end{equation}

```python
joint_probability = df[boolean_mask].shape[0]/df.shape[0]
```

```python
joint_probability
```

# Therefore our final conditional probability: 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous)
\end{equation}

# is given by: 

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous) = \frac{joint probability}{p class equals poissonous}
\end{equation}

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous) = \frac{0.003938946331856229}{0.48202855736090594}
\end{equation}

```python
conditional_probability = joint_probability/p_class_equals_poissonous
```

```python
conditional_probability
```

\begin{equation}
P(cap-shape=x \cap cap-surface=y \cap cap-color=w | class = poissonous) = 0.008171603677221655
\end{equation}

# So, now we are going to compute the following Probability: 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n)
\end{equation}

# let's say that event A = (gill-spacing=c and stalk-surface-below-ring=s), similarly, event B = (class=poissonous and spore-print-color=n)

# Therefore, we can say that: 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n) = P(A|B)
\end{equation}

# Now, according to conditional probability formula, we know that: 

\begin{equation}
P(A|B) = \frac{P(A \cap B)}{P(B)}
\end{equation}

# Therefore, 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n) = \frac{P(gill-spacing=c \cap stalk-surface-below-ring=s \cap class = poissonous \cap spore-print-color=n)}{P(class = poissonous \cap spore-print-color=n)}
\end{equation}

# Now, we have to compute two probabilities: 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s \cap class = poissonous \cap spore-print-color=n)
\end{equation}

# and

\begin{equation}
P(class = poissonous \cap spore-print-color=n)
\end{equation}

# Let's compute the first (numerator) probability: 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s \cap class = poissonous \cap spore-print-color=n) = \frac{Frequency(gill-spacing=c \cap stalk-surface-below-ring=s \cap class = poissonous \cap spore-print-color=n)}{Sample Size(8124)}
\end{equation}

# We know that now, 

# Frequency(gill-spacing=c and stalk-surface-below-ring=s and class = poissonous and spore-print-color=n) = Number of rows in pandas dataframe satisfying all the four conditions. 

# For getting this frequency, therefore, we have to create the boolean mask

```python
df_columns
```

```python
df_columns[13]
```

```python
df_columns[20]
```

```python
boolean_mask = (df['gill-spacing'] == 'c') & (df[df_columns[13]] == 's') & (df[df_columns[0]] == 'p') & (df[df_columns[20]] == 'n')
```

```python
boolean_mask
```

```python
frequency = df[boolean_mask].shape[0]
```

```python
frequency
```

```python
joint_probability_num = frequency/df.shape[0]
```

```python
joint_probability_num
```

# now, similarly I can compute the probability in the denominator: 

\begin{equation}
P(class = poissonous \cap spore-print-color=n) = \frac{Frequency(class=poissonous \cap spore-print-color=n)}{Sample Size(8124)}
\end{equation}

# The frequency above can be computed in the similar manner as before: 

```python
boolean_mask = (df['class'] == 'p') & (df[df_columns[20]] == 'n')
```

```python
joint_probability_denom = df[boolean_mask].shape[0]/df.shape[0]
```

```python
joint_probability_denom
```

# Therefore, the final probability: 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n)
\end{equation}

# is given by 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n) = \frac{joint probability num}{joint probability denom}
\end{equation}

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n) = \frac{0.021664204825209258}{0.027572624322993598}
\end{equation}

```python
conditional_probability = joint_probability_num/joint_probability_denom
```

```python
conditional_probability
```

# Therefore, 

\begin{equation}
P(gill-spacing=c \cap stalk-surface-below-ring=s | class = poissonous \cap spore-print-color=n) = 0.7857142857142858
\end{equation}
